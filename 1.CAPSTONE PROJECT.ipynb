# DATA PREPROCESSING

```python
data['release_date']=pd.to_datetime(data['release_date'])
```

```python
data['release_year'] = pd.DatetimeIndex(data['release_date']).year
data['release_month'] =pd.DatetimeIndex(data['release_date']).month
data['release_day'] = pd.DatetimeIndex(data['release_date']).day
```

```python
all_genres = []
for genre_dict in data['genres']:
   
    try:
        list_represenation = ast.literal_eval(genre_dict)
        genre_list = [ele['name'] for ele in list_represenation]
        all_genres = all_genres + [genre_list]
       
    except:
        genre_list = ['not_found']
        all_genres = all_genres + [genre_list]
```

```python
data['genres']=all_genres
```
